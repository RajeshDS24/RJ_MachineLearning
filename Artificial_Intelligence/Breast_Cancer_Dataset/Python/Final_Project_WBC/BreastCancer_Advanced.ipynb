```python
# Training and testing data
from sklearn.model_selection import train_test_split
from sklearn import svm

from sklearn import metrics
from sklearn.metrics import accuracy_score

import random
```

```python
random.seed(123)
```

```python
# Diving data into train and test

WC_Train, WC_Test = train_test_split(Wiscon_data,test_size = 0.3)
WCtrain_X = WC_Train.drop('diagnosis', axis = 1)
WCtest_X = WC_Test.drop('diagnosis', axis = 1)
WCtrain_Y = WC_Train['diagnosis']
WCtest_Y = WC_Test['diagnosis']
```

```python
# SVM using the default values
SVM_model = svm.SVC()
```

```python
Fit_SVM_model = SVM_model.fit(WCtrain_X, WCtrain_Y)
Fit_SVM_model
```

```python
# Predict SVM model
pred_SVM = Fit_SVM_model.predict(WCtest_X)
```

```python
print(pred_SVM)
```

```python
# Building Confusion Matrix to check the performance
SVM_CM = metrics.confusion_matrix(WCtest_Y, pred_SVM)
print(SVM_CM)
```

```python
accuracy_score(WCtest_Y, pred_SVM)
```

```python
## Implimenting the optimised parameters on raw data.
# Build the SVM classifier

# Import the GridSearchCV method
from sklearn.model_selection import GridSearchCV

SVM_class = GridSearchCV(svm.SVC(gamma ='auto'), {
    'C' : [1,2,5,7,10,12,15,17,20,25,30],
    'kernel' :['rbf','linear']  
} ,cv=10, return_train_score = False )


SVM_class.fit(WCtrain_X, WCtrain_Y)
SVM_class.cv_results_
```

```python
SVM_res = pd.DataFrame(SVM_class.cv_results_)
SVM_res
```
