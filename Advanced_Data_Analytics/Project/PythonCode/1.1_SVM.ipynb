```python
# -*- coding: utf-8 -*-
"""
ICT602 Assignment 2 
Artifact dataset 1
"""


	
from sklearn.datasets import make_blobs
import numpy as np

n_samples = 250
samples, labels = make_blobs(n_samples=n_samples, 
                             centers=([3.5, 3], [6.7, 7.9]), 
                             random_state=0)

newsample = np.array([4.0, 5.86667])
samples = np.vstack((samples, newsample))
  
newlabel = np.array([1])
labels = np.hstack((labels, newlabel))

import matplotlib.pyplot as plt



colours = ('blue', 'red', 'cyan', 'yellow', 'magenta')
fig, ax = plt.subplots()


for n_class in range(2):
    ax.scatter(samples[labels==n_class][:, 0], samples[labels==n_class][:, 1], 
               c=colours[n_class], s=40, label=str(n_class))
    
samples
labels
```

```python
from sklearn import svm
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
```

```python
svm=SVC(kernel = 'rbf',random_state=0,gamma=0.10 ,C=0.001)
svm.fit(samples,labels)
plot_decision_regions(samples,labels,clf=svm) 
plt.legend(loc='upper left')
plt.show()
```
